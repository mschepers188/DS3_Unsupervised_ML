```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
img = plt.imread('img/my_picture.jpeg')
plt.imshow(img)
plt.axis('off');
```

```python
img.shape
```

```python
#create chanels for each color (2 dimensional)
Xred = img[:,:,0]
Xgreen = img[:,:,1]
Xblue = img[:,:,2]
```

```python
def compress(X, n):
    pca = PCA(n_components=n)
    reduced = pca.fit_transform(X)
    recovered = pca.inverse_transform(reduced)
    return recovered
```

```python
n = 25
# run PCA on each channel and combine afterwards in 3dimensional array (picture)
compressed_red = compress(Xred, n)
compressed_green = compress(Xgreen, n)
compressed_blue = compress(Xblue, n)
img_compressed = (np.dstack((compressed_red, compressed_green, compressed_blue))).astype(np.uint8)
```

```python
#viewing the compressed image
plt.imshow(img_compressed)
```
